```python
import pandas as pd

df = pd.read_csv("12_data.csv", parse_dates=["Date"], date_format="%Y-%m-%d %I-%p")
df.head()
```

```python
df.loc[df["Date"] >= "2020"]
```

```python
df.loc[df["Date"] >= "2019" < "2020"]
```

```python
df.loc[(df["Date"] >= "2019") & (df["Date"] < "2020")]
```

```python
df.loc[
    (df["Date"] > pd.to_datetime("2018-12-31"))
    & (df["Date"] < pd.to_datetime("2020-01-01"))
]
```

```python
df.set_index("Date", inplace=True)
df.sort_index(inplace=True)
```
