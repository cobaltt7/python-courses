```python
import pandas as pd

df = pd.read_csv("./1_data.csv", index_col="ResponseId")
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 100)
df.head()
```

```python
india = df[df["Country"] == "India"]
india.head()
```

```python
india.to_csv("./16_india.csv")
pd.read_csv("./16_india.csv", index_col="ResponseId").head()
```

```python
india.to_csv("./16_india.tsv", sep="\t")
pd.read_csv("./16_india.tsv", sep="\t", index_col="ResponseId").head()
```

```python
india.to_excel("./16_india.xlsx")
pd.read_excel("./16_india.xlsx").head()
```

```python
india.to_json("./16_india.json")
pd.read_json("./16_india.json").head()
```

```python
india.to_json("./16_india2.json", orient="records", lines=True)
pd.read_json("./16_india2.json", orient="records", lines=True).head()
```
