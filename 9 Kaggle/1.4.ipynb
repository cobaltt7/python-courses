```python
import pandas

df = pandas.read_csv("./1.3.csv", index_col="name")  # Monday's changes included
df
```

```python
# Drop all rows that are completely identical
# Ignore the `description` column, we can't easily resolve those conflicts so only keep the first
df.drop_duplicates(
    inplace=True,
    keep="first",
    subset=[
        "years",
        "genres",
        "rating",
        "stars",
        "votes",
        "run_time",
        "directors",
    ],
)
```

```python
# Count still-existing `name` duplicates
dupe_counts = df.index.value_counts()
print(dupe_counts[dupe_counts > 1].head(20))
print(sum(dupe_counts[dupe_counts > 1]))
# 60% of the data is duplicated what is this
```

```python
df[df.index.duplicated(keep=False)].sort_index().head(20)
```

```python
# Looks like many individual TV show episodes are included
# We can just leave those as-is
# We should bring back the numerical index if the name isn't going to be unique then
df.reset_index(inplace=True)
```

```python
df["dupe_counts"] = df["name"].map(lambda name: dupe_counts[name])
duplicates = df[df["dupe_counts"] > 1]
duplicates.sort_values(["dupe_counts", "name"]).head(20)
```
