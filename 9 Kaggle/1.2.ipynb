```python
import pandas

df = pandas.read_csv("./1.0.csv", index_col="MOVIES")  # Make the MOVIES field the index
df.head()
```

```python
# Manually rename all columns to use `snake_case`
# Manually rename `ONE-LINE` to `decription`
df.rename(
    columns={
        "YEAR": "years",
        "GENRE": "genres",
        "RATING": "rating",
        "ONE-LINE": "description",
        "STARS": "stars",
        "VOTES": "votes",
        "RunTime": "run_time",
        "Gross": "gross",
    },
    inplace=True,
)
df.index = df.index.rename("name")
df
```

```python
# Default `VOTES` to `0`
df["votes"].fillna(0, inplace=True)
df["votes"]
```

```python
from numpy import NaN

# Replace `\nAdd a Plot\n` in `ONE-LINE` with NaN
df["description"].replace("\nAdd a Plot\n", NaN, inplace=True)
df["description"]
```

```python
# Trim `ONE-LINE` field

df["description"] = df["description"].str.strip()
df["description"]
```

```python
# Drop `Gross` column as it wouldn't be useful enough for anything
df.drop(columns=["gross"], inplace=True)
df
```

```python
# Trim `GENRE` & split on `, `
df["genres"] = df["genres"].str.strip().str.split(", ")
df
```

```python
df.to_csv("./1.2.csv")
```
