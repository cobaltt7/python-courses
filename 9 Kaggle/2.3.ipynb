| What problems should we worry about?                 | What can we do about these problems?                                     |
| ---------------------------------------------------- | ------------------------------------------------------------------------ |
| `Salary Estimate`, `Size`, and `Revenue` are strings | Convert to tuples of numbers                                             |
| `Rating` is included in the `Company Name`           | Parse with string operators                                              |


```python
import pandas

df = pandas.read_csv("./2.2.csv")
df.head()
```

```python
df["company"] = df["company"].apply(lambda info: info.split("\n")[0])
df  # Remove rating data from the company name
```

```python
df
```

```python
df["size"].value_counts()
```

```python
from numpy import NaN


df["size"].replace("Unknown", NaN, inplace=True)
```

```python
def handle_size(size):
    if not isinstance(size, str):
        return size
    sections = size.split()
    return (sections[0], sections[2]) if len(sections) > 2 else (10000,)


df["size"] = df["size"].apply(handle_size)
```
